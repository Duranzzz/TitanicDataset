```python
import pandas as pd
```

```python
titanicData = pd.read_csv('titanic/titanicData.csv')
```

```python
filas, columnas = titanicData.shape
print(f"La cantidad de filas es de: {filas}")
print(f"La cantidad de columnas es de: {columnas}")
```

```python
print("La cantidad de valores nulos por columna son: ")
titanicData.isnull().sum()
```

```python
titanicData.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
```

```python
titanicData['Age'] = titanicData['Age'].fillna(round(titanicData['Age'].mean(), 1))
```

```python
titanicData.dropna(inplace=True)
```

```python
titanicData.to_csv('titanic/titanicData.csv', index=False)
```
